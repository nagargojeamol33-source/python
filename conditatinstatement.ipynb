```python
import speech_recognition as sr
import pyttsx3
import pywhatkit
import datetime
import time

# Voice engine setup (Windows ke liye 'sapi5' best hai)
engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
# Male voice (0), female try karne ke liye 1 ya 2
engine.setProperty('voice', voices[0].id)
engine.setProperty('rate', 150)  # Speed thoda fast

def speak(audio):
    print("Jarvis bol raha: " + audio)
    engine.say(audio)
    engine.runAndWait()

def takeCommand():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Sun raha hu bhai... (bol na)")
        r.pause_threshold = 1.2  # Thoda pause allow
        r.energy_threshold = 300  # Noise adjust
        audio = r.listen(source, timeout=5, phrase_time_limit=10)
    try:
        print("Samajh raha...")
        query = r.recognize_google(audio, language='hi-in')  # Hindi + English mix
        print(f"Tu bola: {query}")
        return query.lower().strip()
    except sr.UnknownValueError:
        speak("Bhai samajh nahi aaya... repeat kar na!")
        return "none"
    except sr.RequestError:
        speak("Internet issue hai boss... Google se connect nahi ho raha.")
        return "none"
    except Exception as e:
        print(e)
        speak("Kuch gadbad ho gayi... phir se bol!")
        return "none"

def get_phone_number():
    speak("Kisko message bhejna hai? Number bol +91 se shuru karke, jaise zero one two three...")
    num_query = takeCommand()
    # Simple clean: digits nikaal aur +91 add agar nahi hai
    digits = ''.join([c for c in num_query if c.isdigit()])
    if len(digits) == 10:
        return "+91" + digits
    elif len(digits) == 12 and digits.startswith("91"):
        return "+" + digits
    else:
        speak("Number sahi nahi samjha... 10 digits ka bol na +91 ke saath")
        return None

def get_message():
    speak("Kya message bhejna hai boss? Bol de full!")
    msg = takeCommand()
    if msg == "none":
        speak("Message nahi mila... phir se try kar")
        return None
    return msg

if __name__ == "__main__":
    speak("Jarvis ready hai Swara bhai! WhatsApp message bhejna hai?")
    speak("Bol 'whatsapp' ya 'message bhej' toh start karte hain. Exit bolne pe bye!")

    while True:
        query = takeCommand()

        if "whatsapp" in query or "message bhej" in query or "bhej" in query:
            speak("Theek hai boss! Shuru karte hain.")
            
            # Phone number le
            phone = None
            attempts = 0
            while phone is None and attempts < 3:
                phone = get_phone_number()
                attempts += 1
            
            if phone is None:
                speak("Number nahi mila... baad mein try karna!")
                continue
            
            # Message le
            message = None
            attempts = 0
            while message is None and attempts < 3:
                message = get_message()
                attempts += 1
            
            if message is None:
                speak("Message nahi mila... cancel kar raha hu!")
                continue
            
            speak(f"Bhej raha hu '{message}' ko {phone} pe... thoda wait kar browser khulne de.")
            
            try:
                # Instant send (wait_time=15 seconds for load)
                pywhatkit.sendwhatmsg_instantly(phone, message, wait_time=20, tab_close=True, close_time=5)
                speak("Message bhej diya boss! Check kar WhatsApp pe.")
                time.sleep(2)  # Thoda pause next command ke liye
            except Exception as e:
                print(e)
                speak("Error aa gaya bhai... shayad WhatsApp Web login nahi hai ya internet slow. Phir se login kar ke try kar!")
        
        elif "exit" in query or "band kar" in query or "bye" in query:
            speak("Theek hai Swara bhai, bye! Jab chahiye Jarvis ko bula lena.")
            break
        
        else:
            speak("Samajh nahi aaya... WhatsApp message ke liye bol 'message bhej' ya 'whatsapp'!")
```
